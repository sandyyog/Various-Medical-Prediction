```python
import pandas as pd
df=pd.read_csv('horse.csv')
df
```

```python
df1=pd.get_dummies(df)
df1=df1.drop_duplicates()
df1
```

```python
x=df1.iloc[:,:].values
x
```

```python
from sklearn.impute import KNNImputer
knn=KNNImputer()
u1=knn.fit_transform(x)
u1

```

```python
from sklearn.preprocessing import MinMaxScaler
obj=MinMaxScaler()
output=obj.fit_transform(u1)
output
```

```python
df2=pd.DataFrame(output)
df2
```

```python
f=df2.iloc[:,0:65].values
f
```

```python
g=df2.iloc[:,69].values
g
```

```python
from sklearn.model_selection import train_test_split
a1,a2,a3,a4=train_test_split(f,g,test_size=0.30,random_state=0)
print(a1,a2,a3,a4)
```

```python
from sklearn.tree import DecisionTreeClassifier
dct=DecisionTreeClassifier(criterion='entropy')
dct.fit(a1,a3)
tree_output=dct.predict(a2)
tree_output
```

```python
from sklearn.metrics import confusion_matrix
log=confusion_matrix(tree_output,a4)
log
```

```python
from sklearn.metrics import confusion_matrix
rc=confusion_matrix(rf_output,a4)
rc
```
